```python
import random
import time

from nebula3.gclient.net import ConnectionPool

from nebula3.Config import Config
from nebula3.common import *

if __name__ == '__main__':
    config = Config()
    config.max_connection_pool_size = 20

    # init connection pool
    connection_pool = ConnectionPool()

    assert connection_pool.init([('172.25.209.218', 9669)], config)


    client = connection_pool.get_session('root', 'nebula')
    url = "Use stockKGQA; MATCH (s:stock)<-[h:hold]-(pof:public_offering_fund)\nWHERE h.position_ratio > 3 \nRETURN s.stock"
    resp = client.execute(url)
    print(resp.error_msg)

    # close connect pool
    client.release()
    connection_pool.close()
```
